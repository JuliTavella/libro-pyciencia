# Tipos de datos: Tuplas

### Definiendo

```python
(1, 'a', 2.3)
```

```python
1, 5, 'xx'
```

```python
t = (5,)
t
```

```python
len(t)
```

```python
t[0]
```

### Unpacking

```python
divmod(60, 7)
```

```python
a, b = divmod(60, 7)
print("a={}, b={}".format(a, b))
```

### Expansor

```python
a, b, c = 1, 2
```

```python
a, b, c = 1, 2, 3, 4, 5, 6
```

```python
a, b, *c = 1, 2, 3, 4, 5, 6
print("a={}, b={}, c={}".format(a, b, c))
```

```python
a, *b, c = 1, 2, 3, 4, 5, 6
print("a={}, b={}, c={}".format(a, b, c))
```

### Construyendo

```python
tuple("foobar")
```

```python
tuple([2, None, 0.1])
```
